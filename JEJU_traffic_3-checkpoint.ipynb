```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
```

```python
df_trf= pd.read_csv('train_jeju.csv')
print(df_trf.shape)
df_trf
```

```python
df_trf.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
str_col = ['day_of_week','start_turn_restricted','end_turn_restricted', 'road_name',\
'start_node_name','end_node_name']
for i in str_col:
    le = LabelEncoder()
    le=le.fit(df_trf[i])
    df_trf[i]=le.transform(df_trf[i])
```

```python
df_trf.head()
```

```python
df_trf.info()
```

```python
df_trf['base_date'] = pd.to_datetime(pd.to_datetime([i for i in df_trf['base_date'].astype('string')]))
```

```python
X_train = df_trf.drop(['base_date','id','target'], axis=1)
y_train = df_trf['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, 
                                            stratify=df_trf['base_hour'], random_state=2021)
```

```python
import lightgbm as lgb
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, KFold
```

```python
#Validation function
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train.values)
    rmse= np.sqrt(-cross_val_score(model, X_train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
model_lgb = lgb.LGBMRegressor(objective='regression',num_leaves=5,
                              learning_rate=0.05, n_estimators=720,
                              max_bin = 55, bagging_fraction = 0.8,
                              bagging_freq = 5, feature_fraction = 0.2319,
                              feature_fraction_seed=9, bagging_seed=9,
                              min_data_in_leaf =6, min_sum_hessian_in_leaf = 11)
```

```python
score = rmsle_cv(model_lgb)
lgb_score = score.mean()
print("LGBMRegressor score: {:.4f} ({:.4f})".format(score.mean(), score.std()))
```

```python
import xgboost as xgb
```

```python
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =7, nthread = -1)
```

```python
score = rmsle_cv(model_xgb)
xgb_score = score.mean()
print("XGBRegressor score: {:.4f} ({:.4f})".format(score.mean(), score.std()))
```
